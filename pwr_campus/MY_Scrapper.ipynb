```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://pwr.nu.edu.pk/cs-faculty/"
r = requests.get(URL, verify = False)
print(r)
```

```python
print(r.content)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
f = soup.findAll('div', class_='team-members')
```

```python
print(f)
```

```python
id_=[]
name=[]
h_edu = []
email=[]
dept = []
desig=[]
approved_phd = []
ext = []
img=[]
profile=[]
```

```python
# print(f[0].find('a')['href'])
# print(f[0].find('li').find('a')['href'])
for faculty in f:
    try:
        id_.append(int(faculty.find('a')['href'].split('=')[-1]))
    except:
        id_.append(-1)

    name.append(faculty.find('h4').text)
    desig.append(faculty.find('h5').text)
    email.append(faculty.find('li').find('a')['href'].split('mailto:')[-1])
    dept.append(faculty.find('a')['href'].split('-')[0])
    img.append('https://pwr.nu.edu.pk/' + faculty.find('img', class_ = 'img-fullwidth')['src'])
    profile.append('https://pwr.nu.edu.pk/' + faculty.find('a')['href'])

for p in profile:
    newres = requests.get(p, verify = False)
    newsoup = BeautifulSoup(newres.content, 'html.parser')
    
    #select second element in list
    newf = newsoup.findAll('div', class_='col-md-8')[1]
    approved_phd.append({True: True, False: False} ["(HEC Approved)" in newf.find('b').text])

    #newsoup object for other properties
    newf = newsoup.findAll('div', class_='tab-content')[0]

    try: h_edu.append(newf.find('div', id = "education").find('ul').find("li").text)
    except: h_edu.append("Not Available")
    
    # print(newf)
    try: ext.append(int(newf.find('div', id = 'contact').find('ul').find("li").text.strip().split(" ")[-1]))
    except: ext.append(-1)

print(h_edu)
```
